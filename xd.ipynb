```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('predicciones_secuencias_allfeatures.csv')
df.head()
```

```python


# Convertir predicted_rating a valores que estan entre 0 a 1 a 0.5 a 5 con paso de 0.5

# Paso 1: Escalar de [0, 1] a [0.5, 5]
scaled_ratings = df['predicted_rating'] * 4.5 + 0.5

# Paso 2: Redondear al múltiplo más cercano de 0.5
df['predicted_rating'] = (scaled_ratings * 2).round() / 2

df.head()
```

```python
# graficar un histograma de la columna predicted_rating
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['predicted_rating'], bins=20, kde=True)
plt.title('Histograma de Predicciones de Rating')
plt.xlabel('Predicted Rating')
plt.ylabel('Frecuencia')
plt.grid()
plt.show()
```

```python
# mostrar el valor máximo y mínimo de la columna predicted_rating
print("Valor máximo de predicted_rating:", df['predicted_rating'].max())
print("Valor mínimo de predicted_rating:", df['predicted_rating'].min())
```

```python
# mostrar los valores únicos de la columna predicted_rating
print("Valores únicos de predicted_rating:", df['predicted_rating'].unique())
```
